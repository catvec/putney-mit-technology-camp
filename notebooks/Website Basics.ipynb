# Website Basics
## HTML Basics
- HTML files tell your browser how the website should look
- This is done via "Tags" and "Styles"

### Tags
- Each "tag" in HTML does something different
  - Each one has a special name in HTML (ex., `p`, `b`, `i`)
- Tags have a start and end
  - Start of tag has the tag's name surrounded by angle brackets:
    ```html
    <name>
    ```
  - End of tag has tag's name surrounded by angle brackets, starting with a slash
    ```html
    </name>
    ```
- Some tags have contents
  ```html
  <p>Tag Contents!</p>
  ```
- Tag examples
  ```html
  <p>Text</p>
  <b>Bold</b>
  <i>Italics</i>
  ```

# HTML Basics
- Tags can be tweaked by specifying “attributes”
- Attributes are values with names
- More common for advanced tags

## Attribute Formatting
- Attributes are only placed in the start of a tag
- Put attributes after the tag name, before the end angle bracket
- Put the attribute value in double quotes

```html
<tagname attribute-name="attribute-value">
```

Only certain tags use attributes.

## Common Attributes
### Link To Another Page

| -------- | -------------- | ----------------- |

# Let's Add Some Style (CSS)
- The whole fun with websites is you can customize them and style them how you want
- You use something called "CSS" for this
- You can change: colors, fonts, sizes, and more!

## What is "CSS"?
- Technically it stands for "Cascading Style Sheets" aka "CSS"
  - Don't think abou this term too much, just know "CSS" helps you change the style of your HTML tags
- CSS is simply a list of **style rules** to apply to your website
  - Each **style rule** has a **selector** which targets which HTML tags you want to change
    - For example you could target all `<a>` tags
  - Each **style rule** sets style **properties** and **values** for the targetted HTML tags to change how they look
    - **Properties** are things like `color`, `width`, `font-size`, ect
    - **Values** depend on the **property**
      - The `color` **property** takes colors as **values** (ex., `blue`, `red`, `purple`)
      - The `width` **property** take sizes in the unit of "pixels" as **values**
        - We use `px` to indicate pixels (ex., `10px` => 10 pixels, `300px` => 300 pixels)
      - There are many many different **properties** each with their own kinds of **values**
      - When in doubt just Google "css make element ..." (ex., "css make element have blue border")
        - Sites like mozilla.org and w3schools.com are good places to look
- Here is how you write a CSS style rule:  
  <img
    alt="Screenshot showing style rule syntax"
    src="https://github.com/catvec/machine-learning-camp/blob/main/assets/website-basics/css-rule-syntax.png?raw=true"
    width="500px"
  />
  - You write the **selector** (In the screenshot the target is all `<p>` tags)
  - Followed by an opening curly brace (`{`)
  - Then a list of **properties** and **values**
    - Write each **property** on a new line
    - Separate the **property** from the **value** with a colon (`:`)
    - Then put the **value** followed by a semicolon (`;`)
  - Finally at the end of the **style rule** there is a closing curly brace (`}`)
    - In the screenshot the **properties** and **values** are:
      | Style Attribute | Style Value |
      | --------------- | ----------- |
      | `width`         | `100px`     |
      | `font-family`   | `Arial`     |

## Advanced Selectors
Up until now we've only talked about fairly simple CSS selectors which target all HTML tags of a certain type. For example the following style rule selects all `<button>` tags.

```css
button {
  // ...
}
```

However, styling all occurances of an HTML tag makes it hard to design a nice looking website. For example, what if you want the "Cancel" button to be red and the "Submit" button to be green? Or what if you want every link that links to Wikipedia to have orange text, but leave all other links untouched?

This is where **classes** and **ids** come in to play. You can target specific occurances of an HTML tag by adding either the `class` or `id` **tag attribute** to the tag. Consider the examples from earlier:

- > A red "Cancel" button and a green "Submit" button
  ```html
  <button id="cancel">
    Cancel
  </button>
  
  <button id="submit">
    Submit
  </button>
  ```
- > Every link to Wikipedia to have orange text, but leave all other links alone
  ```css
  <a class="wikipedia-link"
     href="https://en.wikipedia.org/wiki/Association_football"
  >
    Soccer
  </a>

  is a sport played by many. One prominenant team are the

  <a href="https://www.tottenhamhotspur.com/">
    Tottenham Hotspurs
  </a>

  who used to have a player named

  <a class="wikipedia-link"
     href="https://en.wikipedia.org/wiki/Son_Heung-min"
  >
    Son Heung-min
  </a>
  ```

Notice how the `<button>` tags have `id` attributes with the 2 different values of `cancel` and `submit`. Whereas the `<a>` tags have the `class` attribute with the value of `wikipedia-link`.

In a CSS **style rule** the selector can target one HTML tag that has an `id` attribute by putting a hashtag (`#`) and then the `id` value. For example:

```css
#cancel {
  color: red;
}

#submit {
  color: green;
}
```

These 2 CSS **style rules** make the HTML tag with the `id="cancel"` attribute have red text and the HTML tag with the `id="submit"` attribute have green text.

To make a CSS **style rule** which targets all HTML tags with a specific `class` attribute value put a dot (`.`) followed by the `class` value. For example:

```css
.wikipedia-link {
  color: orange;
}
```

## When to use id vs class?
Both `id` and `class` can be used to target more specific HTML tags for styling. However, there is a rule which helps you decide which HTML attribute to use:

Only **one** HTML tag in the entire page can have a specific `id` value. You cannot have 2 HTML tags which both have the same `id` attribute value.

This makes `id` best for styling one specific HTML tag. If you need to style multiple HTML tags use `class`.

- **Illegal, do not do:**
  ```html
  <button id="cool-button">
    Press me!
  </button>

  <button id="cool-button">
    No! Press me!
  </button>
  ```
  This is not allowed because both buttons have the `id="cool-button"` attribute.
- **Instead use a class:**
  ```html
  <button class="cool-button">
    Press me!
  </button>

  <button class="cool-button">
    No! Press me!
  </button>
  ```
