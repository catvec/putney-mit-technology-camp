## Dictionaries
Dictionaries allow you to store any sort of value under a label. These can be convenient because you can store a collection of values all in one variable.

To define a dictionary:

- Start with an opening curly brace `{`
- Put a label as a string (example: `"foo"`)
- Put a colon `:`
- Then the value
- Followed by a comma
- You can do this as many times as you need

````python
# Example of making a dictionary
bob = {
    "first name": "Bob",
    "last name": "Bobbert",
    "age": 87,
    "favorite color": "blue",
    "pets": ["sally", "will", "tony"],
}
````

Here you can see we defined a variable named `bob`. This variable is a dictionary. The dictionary has several values with names:

| Label | Value |
| - | - |
| first name | Bob |
| last name | Bobbert |
| age | 87 |
| favorite color | blue |
| pets | sally, will, tony |

Similar to lists, you can access a value in a dictionary based on its label:

- Put the variable name
- Then between square brackets `[` `]` put the label of the value as a string

For example to access the age of bob:

````python
# Access the age of bob
bob = {
    "first name": "Bob",
    "last name": "Bobbert",
    "age": 87,
    "favorite color": "blue",
    "pets": ["sally", "will", "tony"],
}

print("Bob is", bob["age"])
````

You can also set the value by its label in a dictionary. For example, if it's Bob's birthday and we need to make him one year older:

````python
# Change the value of age
bob = {
    "first name": "Bob",
    "last name": "Bobbert",
    "age": 87,
    "favorite color": "blue",
    "pets": ["sally", "will", "tony"],
}

bob["age"] = 88
````

# Comparison
Programs need a way of asking questions about the data they have. To do this they make comparisons between two values.

For example a computer might want to know:

- Is a number if bigger than another, or if a number is equal?
- Does a string has a certain word in it?

In Python we write comparisons by putting a value, then an comparison operator, then another value:

```
<value a> <comparison operator> <value b>
```

The value on the left (`<value a>`) and the value on the right (`<value b>`) can be anything. A variable, a value, both work.

Some common comparison operators are:

| Operator | What it Does             | Usable on Types | Example                  |
| -------- | ------------------------ | ---------------- | ------------------------ |
| `>` | Greater than | Numbers | `2 > 3` is `True`  |
| `<` | Less than | Numbers | `2 < 3` is `False` |
| `>=` | Greater than or equal to | Numbers | `3 >= 3` is `True` |
| `<=` | Less than or equal to | Numbers | `3 <= 3` is `True` |
| `==` | Equal to | Most Anything | `3 == 3` is `True` |
| `in` | Contained by | Strings and List | `"a" in "fan"` is `True` |

Notice that certain operators only work on numbers. Where others (like `==`) work on almost any type of data.

Here are some more examples of comparison expressions:

````python
# Is 3 greater than 3?
3 > 3
````

````python
# Is 3 less than 3?
3 < 3
````

````python
# Is 3 greater or equal to 3?
3 >= 3
````

````python
# Is 3 less than or equal to 3?
3 <= 3
````

````python
# Is the number of cows greater than the bales of hay?
number_of_cows = 12
bales_of_hay = 6

number_of_cows > bales_of_hay
````

````python
# Is a equal to b?
a = 6
b = a + 2

a == b
````

````python
# Is 4 in this list?
4 in [1, 2, 3]
````

````python
# Is the "miss" in "mississippi"?
"miss" in "mississippi"
````

Notice how all comparison expressions result in either a `True` or `False` result. This is important for a computer. Answers need to be clear.

# If Statements
One of the most useful things a computer can do is look at a variable and make a decision. For example at a bank a program may look at your balance and only allow you to withdraw money if you have enough. Or a Google Docs might only let you edit a document if you are the owner.

We use comparisons, which we just learned about, to write if statements:

- First write the word `if`, this tells Python that your program is about to make a decision
- Then immediately after `if` put a space and then a comparison
  - For example:
    ```python
    if 800 > 9:
    ```
  - Any of the comparisons we learned above work in an if statement
- Immediately after the comparison put a colon `:`
  - For example:
    ```python
    if 800 > 9:
    ```
- Then on the following line put 4 spaces and then any code you want to run
  - If you want to write code that takes multiple lines make sure to add 4 spaces before each line
  - This process of putting 4 spaces at the beginning of a line is called **indenting**
  - For example:
    ```python
    if 800 > 9:
        print("I will only run if 800 is greater than 9")
    ```

Let's take a look at some examples:

````python
# We will only print if 800 is greater than 9
if 800 > 9:
    print("hello")
````

````python
# Only print a congradulations if more than an hour was spent reading
minutes_reading = 120
if minutes_reading > 60:
    print("Congradulations you read for more than a hour")
````

````python
# Only say the guess was correct if the guess matched the secret number
guess = 12
secret_number = 10
if guess == secret_number:
    print("You guessed correctly")
````

Sometimes you want to do something if the comparison was `True`, but you may also want to do something **else** if the the comparison was `False`.

For example imagine you are making a program for a door lock. If the person enters the code correctly you want to unlock the door. But if the person doesn't enter the code correctly you may want to alert security.

To do something when the comparison doesn't result in `True` you can write `else`:

- First write your `if` statement as you normally would:
  - For example:
    ```python
    if 8 > 700:
        print("Somehow 8 is greater than 700")
    ```
- Then on the next line write `else:` without any spaces (aka indentation) at the beginning of the line
  - For example:
    ```python
    if 8 > 700:
        print("Somehow 8 is greater than 700")
    else:
    ```
- Finally, on the next line write the code you want to run if the comparison **does not** equal `True`
  - Make sure to add 4 spaces at the beginning of this line and any other lines
  - For example:
    ```python
    if 8 > 700:
        print("Somehow 8 is greater than 700")
    else:
        print("Of course 8 is not greater than 700 :/")

Let's take a look at some examples:

````python
# Alert the guards if the door code is not entered correctly
door_code = 3849
user_input = 8890

if door_code == user_input:
    print("Welcome home, unlocking...")
else:
    print("Security has been alerted, please leave the premises")
````

````python
# Allow money to be withdrawn from a user's bank account, or show a message to a user if they don't have enough
account_balance = 1_000_000
purchase_price = 10

if account_balance - purchase_price > 0:
    account_balance = account_balance - purchase_price
    print("Purchase approved, you have", account_balance, "left in your account")
else:
    print("Insufficient funds")
````

# Loops
Loops allow you to do the same code more than once. For example imagine you have a list of grades from a class and you want to figure out how many A's people got. The best way to do it is a loop!

Loops are used only with lists.

To write a loop:

- Write the word `for`
- Followed by a space and then the name of a variable which will hold the current value for the loop
  - For example:
    ```python
    for fruit
    ```
- Then put a space followed by `in` and another space:
  - For example:
    ```python
    for fruit in
    ```
- Next put a list that you wish to loop over followed by a colon
  - You can put just a list here:
    ```python
    for fruit in ["apple", "banana", "cherry"]:
    ```
  - Or you can put a variable which is a list:
    ```python
    fruit_list = ["apple", "banana", "cherry"]
    
    for fruit in fruit_list:
    ```
- Finally, similar to an if statement, on the next line put 4 spaces and then the code which you want to run for each item of the list

For example:



````python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print("The current fruit is:", fruit)
````

Notice that we were able to print every item in the list, while only writing `print` once? This is the power of lists. They let you do repetitive things without repeating yourself.

Here are some more examples:

````python
# Add 3 to each item in the list
numbers = [1, 2, 3, 4]

for num in numbers:
    print(num + 3)
````

````python
# Count how many numbers in the list are greater than 3 when divided by 2
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
count_over_3 = 0

for num in numbers:
    if num / 2 > 3:
        count_over_3 = count_over_3 + 1

print("There were", count_over_3, "numbers greater than 3 when divided by 2")
````

# Functions
Loops are great for doing repetitive things. But they only work for lists. For other actions that you want to repeat frequently you can use **functions**.

- Functions let you reuse code
- Functions have names
- Functions take arguments
  - Arguments are like variables which only exist within the function
  - Arguments let you customize the behavior of the code within the function
- You've already seen a function: `print(message)`
- Define a function:
  ```python
  def function_name(argument_1, argument_2):
      print(argument_1)
      print(argument_2)
  ```
  - Write the word `def`
  - Followed by the name of the function
  - Function names must meet the following rules:
    - They must start with a lowercase letter
    - They can only contain lowercase letters, numbers, and underscores
    - No spaces are allowed in function names
    - For easier readability use underscores to separate words (ex., `my_name_is`)
  - Then put an opening parenthesis `(`
  - Followed by the names of your arguments, each argument separated by a comma
  - Then a closing parenthesis `)` and a colon `:`
  - On the following lines put the code that should run when the function is run
    - Code inside a function should be have 4 spaces before it

For example let's make a function named `greet`. It will take two arguments: `first_name` and `last_name`. The function will use `print()` to say hello to the person with the specified name

````python
# Greet function
def greet(first_name, last_name):
    print("Hello", first_name, last_name)

greet("Lewis", "Carroll")
````

````python
# Here is an example where we define a function which always multiplies the input number by 2
def multiply_times_2(number):
    print(number, "multiplied by 2 is:", number * 2)

multiply_times_2(10)
````

## Functions Can Return Values
These functions are all cool and nice, but not incredibly useful. What if you want to compute some complicated value inside a function, and then use the result somewhere else?

For this we can **return** a value from a function. To do this inside a function simply type `return` followed by the value you wish to return. When you use the function you can then assign the returned value to a variable, or just use the value right then and there.

For example, let's revisit the `multiply_times_2` function. But this time let's return the value:

````python
def multiply_times_2(number):
    return number * 2

result = multiply_times_2(10)
print(result)
````

See how instead of printing `number * 2` we actually `return` the value? Then we call `multiply_times_2` and assign the returned value to a variable named `result`.

You can also just treat the function as the value which was returned. For example:

````python
def multiply_times_2(number):
    return number * 2

print("10 times 2 plus 6 is:", multiply_times_2(10) + 6)
````

Here we didn't even assign the returned value from `multiply_times_2` to a variable. Instead we just used it right there in `print()`.

# Classes
Before we start making our own AI models we need to learn one last important concept.

Up until now we've only been defining functions in a simple way. And we have to pass all the variables we want to these functions. This works well for a lot of cases. But imagine we have several functions which all need the same variables and all related to each other.

For this we can use classes. In Python classes are a convenient way to group variables and functions together.

Python classes also let you define logic in a re-usable way. Imagine you have some variables and functions which track the money in a user's bank account. You could write some Python code like:



````python
# Bank account example
def buy(money, amount):
    money_after_purchase = money - amount
    if money_after_purchase < 0:
        print("Sorry you don't have enough money")

    print("You bought something for", amount, "and now you have", money_after_purchase)
    return money_after_purchase

def deposit_money(money, amount):
    money_after_deposit = money + amount
    print("You put", amount, "in your bank account and now have", money_after_deposit)
    return money_after_deposit

my_money = 100
my_money = buy(my_money, 10)
my_money = deposit_money(my_money, 100)
````

This works well enough, but it has two major problems:

- You keep needing to pass `my_money` to every function
- This doesn't work very well if we want to track multiple people's bank accounts
  - We'd have to keep making new variables like `bobs_money` and `susans_money`
  - We'd have to make sure to pass the correct variable in each time
  - Imagine if we accidentally took money out of Susan's account for a purchase bob made:
    ```python
    bobs_money = 100
    susans_money = 100

    susans_money = buy(bobs_money, 30)
    ```

In a moment we'll learn how we can make our own classes. But let's first take a look at what this bank account example would look like with classes:

_(This example won't run, we're just showing what the code could look like)_
```python
bobs_account = Account(100)
susans_account = Account(100)

bobs_account.buy(30)
susans_account.buy(10)

susans_account.deposit_money(300)
```

Notice how much easier this looks. No longer do we have to pass the amount of money each user has to every function. And we don't have to worry about mixing up who's account is who's when making purchases or deposits.

Now let's take a look at how we can make a class. Let's start by defining a class:

```python
class Account:
```
First to make a class type the word `class`, then a space, and then the name of your class, followed by a colon (`:`)

- Class names must start with a capital letter, followed by lowercase letters and numbers
- If you want your class name to have more than one word make sure you capitalize the first letter of the word (ex., to name your class "bank account" you would name it `BankAccount`)

Now let's set some variables inside the class. These variables will be separate for each copy of a class you make. So the variables in `bobs_account` are separate from the variables in `susans_account`:

```python

    def __init__(self, starting_money):
        self.money = starting_money
```

- To set up variables in a class you need to define a function inside the class named `__init__`
  - This is a special name which Python will use to set up your class
- Every function in a class must have its first argument be named `self`
  - This is used to refer to the copy of the class
  - With it you can access variables and even functions which the class has
- To set a variable in our class you just put `self.` before the variable name
- All functions in a class have 4 spaces at the start of each line
- Notice how the `__init__` function can take its own arguments too
  - In this case we are taking the initial value of a users bank account

Now let's define some more functions in our class:

        if money_after_purchase < 0:

        self.money = money_after_purchase

# Class Inheritance
Classes have one advantage which makes them incredibly useful. When defining a class, you can use another class as a template.

This is best shown with an example. Here we have a class named `Animal`:

````python
# Animal class, it has a food variable which is changed by eating and running
class Animal:
    def __init__(self):
        self.food = 0

    def eat(self):
        self.food = self.food + 1
        print("After eating my food is: ", self.food)

    def run(self):
        if self.food < 1:
            print("I don't have enough food to run")
        else:
            self.food = self.food - 1
            print("After running my food is: ", self.food)
````

Now imagine we want to define 2 classes: `Dog` and `Cat`. And we want each class to have `food`, `eat()`, and `run()`.

We could just copy paste all that code again. But, there is a better way. We can use the `Animal` class as a template for our `Dog` and `Cat` classes.

To do this we put the class we want to template from in parenthesis after our class name:

```python
class NewClassName(TemplateClassName):
```

- Put your new class name:
  ```python
  class NewClassName
  ```
- Then put the name of the class you want to template from inside parenthesis:
  ```python
  class NewClassName(TemplateClassName):
  ```

Let's see what this looks like with `Animal` as the template class and `Dog` and `Cat` as the new classes:

````python
# Dog uses Animal as a template
class Dog(Animal):
    def bark(self):
        print("woof! woof!")

# Cat uses Animal as a template
class Cat(Animal):
    def meow(self):
        print("meow! meow!")

# We can use the Animal class functions!
mark = Dog()
gram = Cat()

mark.bark()
mark.eat()
mark.run()
mark.eat()

gram.meow()
gram.eat()
gram.run()
````

You can see that both our `Dog` and `Cat` classes use `Animal` as a template. And that all the functions from `Animal` are available in `Dog` and `Cat`. Just like normal you can define methods in `Dog` and `Cat` which aren't in `Animal`.

We also call this behavior "inheriting" in programming. We say `Dog` and `Cat` **inherit** from `Animal`. We also say `Animal` is the **parent** class while `Dog` and `Cat` are **child** classes.

This is just different termonology for the same thing.

## Overriding
What if we want to get all the methods from the `Animal` class, but for the `Cat` class we want to modify how the `run()` function works? Maybe because cats are smaller and use less food when running.

Doing this is actually very straight forward, we just define the method we want to override in our `Cat` class. Then when we call `run()` on the `Cat` class it will use our new `run()` method instead:

````python
# Cat has a custom run() function
class Cat(Animal):
    def meow(self):
        print("meow! meow!")

    def run(self):
        # Special run function for cats
        if self.food < 0.5:
            print("I am a cat and I don't have enough food to run")
        else:
            self.food = self.food - 0.5
            print("I am a cat and after running my food is: ", self.food)

gram = Cat()

gram.meow()
gram.eat()
gram.run()
gram.eat()
````

Great! You can see that we wrote our own custom `run()` function. And now when our cat runs it only uses `0.5` food instead of `1`.

## Extending
But what if instead of overriding a function we want to keep its original behavior and also add our own on top of it?

For example what if in our `Dog` class we want to make the `run()` function print `"woof woof woof bark"` and also decrease the food.

This is also super easy in Python. Inside of our `Dog` class we define a function named `run()`. Doing this by itself would completely override `run()` so we would loose the `run()` functionality from the `Animal` class.

So what we do is inside of the `Dog` `run()` function we call `super().run()`. This might lookk a bit weird, but calling `super().function_from_parent_class()` will call the logic from the parent class (`Animal`):

````python
# Dog has a custom run() function
class Dog(Animal):
    def bark(self):
        print("woof! woof!")

    def run(self):
        # Special run function for dogs
        # First we call the Animal.run() function, so that our Dog uses food while it runs
        super().run()

        # Then we write our custom code after
        print("woof woof woof bark")

mark = Dog()

mark.eat()
mark.run()
mark.eat()
````

You can see that calling the `run()` function of `Dog` still uses food, but in addition it also prints our custom dog running message: `"woof woof woof bark"`.

This is because we called `Animal`'s `run()` function inside of the custom `Dog` `run()` function via the special `super().run()` call.
